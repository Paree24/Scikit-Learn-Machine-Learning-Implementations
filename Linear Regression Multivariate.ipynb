```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression 
```

```python
mydat=pd.read_csv(r"C:\Users\PareeKatti\Ml_Data\Advertising.csv")

mydat.head(1)

X_Adinput=mydat.iloc[:,1:4]

X_Adinput.size

Y_Adoutput=mydat.iloc[:,4:]

Y_Adoutput.size

X_Adtrain=X_Adinput[:175]
X_Adtest=X_Adinput[175:]
Y_Adtrain=Y_Adoutput[:175]
Y_Adtest=Y_Adoutput[175:]

X_Adtrain

teacher=LinearRegression()
learner1=teacher.fit(X_Adtrain,Y_Adtrain)

Y_Result=learner1.predict(X_Adtest)

Y_Adtest.shape

X_Adtest.shape

Y_Result.shape
```

```python
plt.plot(X_Adtest['TV'],Y_Adtest,'ro')
plt.plot(X_Adtest['TV'],Y_Result,'bo')
plt.xlabel("X_test")
plt.ylabel("Y_test")
plt.legend(["Ya","Yp"])
plt.show()

plt.plot(X_Adtest['radio'],Y_Adtest,'ro')
plt.plot(X_Adtest['radio'],Y_Result,'bo')
plt.xlabel("X_test")
plt.ylabel("Y_test")
plt.legend(["Ya","Yp"])
plt.show()

plt.plot(X_Adtest['newspaper'],Y_Adtest,'ro')
plt.plot(X_Adtest['newspaper'],Y_Result,'bo')
plt.xlabel("X_test")
plt.ylabel("Y_test")
plt.legend(["Ya","Yp"])
plt.show()
```
