# Training and Predicting Using K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
teacher=KNeighborsClassifier(n_neighbors=31)
```

```python
learner=teacher.fit(inputtrain,outputtrain)
```

```python
predictedoutput=learner.predict(inputtest)
```

# Measuring Accuracy

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(outputtest,predictedoutput)*100
```
